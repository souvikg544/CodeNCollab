```python
df2['Income']=[700,500,600,400,800,1000]
```

```python
df2.head()
```

```python
plt.figure(figsize=(12,8))
plt.bar(df2['Country'],df2['GDP'],width=0.3,color='red')
plt.bar(df2['Country'],df2['Income'],align='edge',width=0.3,color='green')
plt.xlabel("Country")
plt.ylabel("GDP")
plt.legend()
plt.show()
```
