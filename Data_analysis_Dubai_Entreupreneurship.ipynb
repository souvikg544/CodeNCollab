```python
def compute_category(x):
  s=x.split(',')[0].lower().strip()
  if(s=='other'):
    try:
      s=x.split(',')[1].lower().strip()
    except:
      s='other'
  return s
```

```python
df['roles'].unique()
```

```python
df['category']=df['category_groups_list'].apply(compute_category)
```

```python
len(df['category'].unique())
```

```python
df['category'].value_counts()
```

```python
df.drop(columns=['category_groups_list'],inplace=True)
```

```python
df['employee_count'].value_counts()
```

```python
def employee_count(s):
  if(s=='10000+'):
    return 10000
  elif(s=='unknown'):
    return 0
  else:
    return int(s.split('-')[1])-int(s.split('-')[0])
```

```python
df['employee_count']=df['employee_count'].apply(employee_count)
```

```python
c=['raised_amount_usd','post_money_valuation_usd','investor_count']
for c in c:
  df1[c].fillna(df1[c].mean(),inplace=True)
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] =df1.columns
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(df1.values, i) for i in range(len(df1.columns))]
vif_data["mean"]=vif_data["feature"].apply(lambda x: np.mean(df1[x]))
vif_data["S.d"]=vif_data["feature"].apply(lambda x: np.std(df1[x]))
vif_data["50th quantile"]=vif_data["feature"].apply(lambda x: np.quantile(df1[x], .50))
vif_data["25th quantile"]=vif_data["feature"].apply(lambda x: np.quantile(df1[x], .25))
vif_data["75th quantile"]=vif_data["feature"].apply(lambda x: np.quantile(df1[x], .75))
vif_data["max"]=vif_data["feature"].apply(lambda x: np.max(df1[x]))
vif_data["min"]=vif_data["feature"].apply(lambda x: np.min(df1[x]))
```
